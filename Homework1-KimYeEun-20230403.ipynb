```python
#6
import matplotlib.pyplot as plt
%matplotlib inline

class Student:
    def __init__(self,age,gender,score):
        self.age = age
        self.gender = gender
        self.score = score
    
    def educlev(self): 
        if self.age < 8:
            print("Preschool")
        elif self.age > 7 and self.age < 14:
            print("Elementary school student")
        elif self.age > 13 and self.age < 17:
            print("Middle school student")
        elif self.age > 16 and self.age < 20:
            print("High school student")
        else:
            print("Undergrad")
        
    def grades(self):
        if self.score > 94:
            print("A+")
        elif self.score > 89:
            print("A")
        elif self.score > 84:
            print("B+")
        elif self.score > 79:
            print("B")
        elif self.score > 74:
            print("C+")
        elif self.score > 69:
            print("C")
        elif self.score > 64:
            print("D+")
        elif self.score > 59:
            print("D")
        else:
            print('F')

    
    def definegen(self):
        if self.gender == "M" or self.gender == 'm':
            print("Male")
        elif self.gender == "F" or self.gender == 'f':
            print("Female")
        else:
            print("Non-binary")
        
        
#Example            
#1)
Steven = Student(7, "M", 88.4)
Steven.educlev()
Steven.grades()
Steven.definegen()

#2)
Julie = Student(15, "F", 73)
Julie.educlev()
Julie.grades()
Julie.definegen()

#3)
Kim = Student(19, "n", 91)
Kim.educlev()
Kim.grades()
Kim.definegen()
```
